```python
### なぜ列にアクセスできない？
覚え方：
df.loc[行, 列] - 行と列の両方を指定
df['列名'] - 列のみを指定
df.loc['行名'] - 行のみを指定
```

```python
df.loc['col1']
```

```python
df['col1']
```

### なぜ行にだけloc関数が必要？
DataFrameでは「列を取り出す」操作が最も頻繁
だからdf['列名']というショートカットが用意されている
行の取り出しは相対的に少ないので、明示的にlocを使う設計

```python
df.loc[[1,3], ['col1','col3']]
```

```python
df
```

```python
df.loc[[1,3], :]
```

```python
df.loc[[1], ['col1']]
```

```python
df[[:], ['col3']]
```

### これはなぜダメ？ df[[:], ['col3']]

df[] の中は列の指定のみ
[:] はスライス記法でリスト内では使えない
df[] に複数の引数は渡せない

覚え方：
df[] = 列専用
df.loc[] = 行・列両方OK


```python
# これなら指定した列だけ出力できる
df['col3']
```
